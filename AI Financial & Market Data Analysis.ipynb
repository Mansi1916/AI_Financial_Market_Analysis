```python
#How much amount the companies spent on R & D

df.head()
```

```python
print("Company's spending for R & D in billion dollors")
RD = df.groupby('Company')['R&D_Spending_USD_Mn'].sum()/1000
```

```python
#Drawing a Bar Plot to show the amount spent on R & D by the companies
plt.bar(RD.index, RD.values, color = ['cyan' , 'black' ,'magenta'])
plt.title("R&D Spending by Companies")
plt.xlabel("Company")
plt.ylabel("Company's spending for R & D in billion dollars")

plt.show() 
```

```python
#Revenue Earned by the Companies
print("Revenue Earned by the Companies")
```

```python
print("AI_Revenue_USD_Mn")
rev = df.groupby('Company')['AI_Revenue_USD_Mn'].sum()/1000
rev
```

```python
plt.bar(rev.index ,rev.values, color = ['pink' , 'black' ,'red'] , width = 0.3)
plt.title('Revensue in AI earned by the Companies')
plt.xlabel('Company')
plt.ylabel('Amount in USD_$Bn')
plt.show()
```

```python
# bar plots to show expenditure & revenue of the companies

plt.figure(figsize = (10,4))

plt.subplot(1,2,1)
plt.bar(RD.index, RD.values, color = ['cyan' , 'black' ,'magenta'])
plt.title("R&D Spending by Companies")
plt.xlabel("Company")
plt.ylabel("Company's spending for R & D in billion dollars")


plt.subplot(1,2,2)
plt.bar(rev.index ,rev.values, color = ['pink' , 'black' ,'red'] , width = 0.3)
plt.title('Revensue in AI earned by the Companies')
plt.xlabel('Company')
plt.ylabel('Amount in USD_$Bn')

plt.show()
```
